```python
from scipy.stats import f_oneway
group1 = train[train['Sex'] == 'male']['Fare']
group2 = train[train['Sex'] == 'female']['Fare']

anova_result = f_oneway(group1, group2)

# Significance level
alpha = 0.05

print("ANOVA Result:")
print("F-statistic:", anova_result.statistic)
print("p-value:", anova_result.pvalue)

if anova_result.pvalue < alpha:
    print("The difference in means is statistically significant (reject null hypothesis)")
else:
    print("The difference in means is not statistically significant (fail to reject null hypothesis)")
```

```python
from scipy.stats import f_oneway
group1 = test[test['Sex'] == 'male']['Fare']
group2 = test[test['Sex'] == 'female']['Fare']

anova_result = f_oneway(group1, group2)

print("ANOVA Result:")
print("F-statistic:", anova_result.statistic)
print("p-value:", anova_result.pvalue)

# Significance level
alpha = 0.05

if anova_result.pvalue < alpha:
    print("The difference in means is statistically significant (reject null hypothesis)")
else:
    print("The difference in means is not statistically significant (fail to reject null hypothesis)")
```


UNIVARIATE GRAPHICAL : Means dealing with single variable graphically


```python
#HISTOGRAM

sns.histplot(train['Age'], kde=True, color='#CB4335',edgecolor='black',alpha=0.3,linewidth=1)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#VIOLINPLOT
sns.violinplot(x='Fare', data=train,inner='quartile',linewidth=1.5,edgecolor='#2980B9 ',color='#7FB3D5')

plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
colors = ['#F4D03F', '#CD6155']

plt.figure(figsize=(15, 10)) 
sns.countplot(x='Survived', data=train, palette=colors, edgecolor='black')
plt.title('Distribution of Survived',fontsize=18)
plt.xlabel('Survived',fontsize=14)
plt.ylabel('Frequency',fontsize=14)

plt.xticks(ticks=[0, 1], labels=['Not Survived', 'Survived'], fontsize=12)
plt.yticks(fontsize=12)

total_count = len(train['Survived'])  # Total count of observations
categories = ['Not Survived', 'Survived']  # Define categories in the desired order
for i, category in enumerate(categories):
    count = (train['Survived'] == i).sum()  # Count occurrences of each category
    percentage = (count / total_count) * 100  # Calculate percentage
    plt.text(i, count + 10, f'{count} ({percentage:.1f}%)', horizontalalignment='center', fontsize=16)

plt.show()
```

```python
color=['#C39BD3']
sns.boxplot(x='SibSp', data=train, palette=color,linewidth=1.5, fliersize=8)
plt.title('Distribution of SibSp',fontsize=18)
plt.xlabel('SibSp',fontsize=14)
plt.ylabel('Frequency',fontsize=14)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set custom color palette
palette = ['#AED6F1', '#ABEBC6', '#F9EBEA']

# Create count plot with enhanced features
plt.figure(figsize=(12, 8))  # Set figure size
sns.countplot(x='Embarked', data=train, palette=palette, edgecolor='black', linewidth=1)
plt.title('Distribution of Embarked', fontsize=18)
plt.xlabel('Embarked', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=14)
plt.show()
```

```python
survival_rate = train.groupby('Sex')['Survived'].mean().reset_index()
palette = {'male': '#FDEDEC', 'female': '#A9DFBF'}
plt.figure(figsize=(8, 6))
sns.barplot(x='Sex', y='Survived', data=train,palette=palette,edgecolor='black',linewidth=1.5)
plt.title('Survival Rate by Gender',fontsize=18)
plt.xlabel('Gender',fontsize=14)
plt.ylabel('Survival Rate',fontsize=14)
plt.ylim(0, 1)  
```

```python
palette = {'male': '#A9CCE3', 'female': '#E6B0AA'}
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Parch', y='Fare', data=train,hue='Sex',edgecolor='black',s=50,linewidth=0.5,palette=palette)
plt.title('Scatterplot of Fare vs Parch')
plt.xlabel('Parch')
plt.ylabel('Fare')
plt.tick_params(axis='both', which='major', labelsize=14)
```

```python
palette = {'male': '#1A5276', 'female': '#6C3483'}
plt.figure(figsize=(10, 8))
sns.jointplot(x='Age', y='Fare', data=train, kind='kde',edgecolor='black',linewidth=1.5,palette=palette, hue='Sex', hue_order=['male', 'female'], alpha=0.7, fill=True)
plt.suptitle('Jointplot of Age vs Fare', y=1.02) 
plt.show()
```

```python
gender_counts = train['Sex'].value_counts()
plt.figure(figsize=(10, 8))
wp={'linewidth':1,'edgecolor':'black'}
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%',colors=['#7FB3D5', '#FDEDEC'],shadow=True,frame=True,wedgeprops=wp)
plt.title('Distribution of Passengers by Gender',fontsize=16)
plt.legend(loc='upper right', fontsize=14)
plt.show()
```

```python
palette = {'male': 'skyblue', 'female': 'salmon'}
plt.figure(figsize=(12, 8))
sns.swarmplot(x='Sex', y='Age', data=train, hue='Sex', edgecolor='black',linewidth=1.5,palette=palette, dodge=True, size=8)
plt.title('Swarmplot of Age by Gender', fontsize=18)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Age', fontsize=14)
plt.legend(title='Gender', fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=14)
```
